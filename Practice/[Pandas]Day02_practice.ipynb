```python
# Q20. Îç∞Ïù¥ÌÑ∞Î°úÎìú
#-------------------------------------
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv')
```

```python
# Q21. quantityÏª¨Îüº Í∞íÏù¥ 3Ïù∏ Îç∞Ïù¥ÌÑ∞Î•º Ï∂îÏ∂úÌïòÏó¨ Ï≤´ 4ÌñâÏùÑ Ï∂úÎ†•
#-------------------------------------
Q21 = df.loc[df['quantity'] == 3].head(4)
Q21
```

```python
# Q22. quantityÏª¨Îüº Í∞íÏù¥ 3Ïù∏ Îç∞Ïù¥ÌÑ∞Î•º Ï∂îÏ∂úÌïòÏó¨ 
#      indexÎ•º 0Î∂ÄÌÑ∞ Ï†ïÎ†¨ÌïòÍ≥† Ï≤´ 6ÌñâÏùÑ Ï∂úÎ†•
#-------------------------------------
Q22 = df.loc[df['quantity']==3].reset_index(drop=True).head(6)
Q22
```

```python
# Q23. quantity , item_price ÎëêÍ∞úÏùò Ïª¨ÎüºÏúºÎ°ú Íµ¨ÏÑ±Îêú
#      ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùÑÏÉùÏÑ±
#-------------------------------------
Q23 = df[['quantity','item_price']]

Q23
```

```python
# Q24. item_price Ïª¨ÎüºÏùò Îã¨Îü¨ÌëúÏãú Î¨∏ÏûêÎ•º Ï†úÍ±∞ÌïòÍ≥†
#      float ÌÉÄÏûÖÏúºÎ°ú Ï†ÄÏû•ÌïòÏó¨ new_price Ïª¨ÎüºÏóê Ï†ÄÏû•ÌïúÎí§
#      ÏïûÏùò 5Í∞úÏ∂úÎ†•
#-------------------------------------
df['new_price'] = df['item_price'].str[1:].astype(float)
df['new_price'].head()
```

```python
# Q25. new_price Ïª¨ÎüºÏù¥ 5Ïù¥ÌïòÏùò Í∞íÏùÑ Í∞ÄÏßÄÎäî Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Ï∂îÏ∂ú
#-------------------------------------
Q25 = df.loc[df['new_price'] <= 5]
cnt = len(df.loc[df['new_price'] <= 5])
display(Q25)
print(cnt)
```

```python
# Q26. item_nameÎ™ÖÏù¥ Chicken Salad Bowl Ïù∏ Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùÑ Ï∂îÏ∂úÌïòÍ≥†
#      index Í∞íÏùÑ Ï¥àÍ∏∞Ìôî
#-------------------------------------

Q26 = df.loc[df['item_name'] == 'Chicken Salad Bowl'].reset_index(drop=True)
Q26 
```

```python
# Q27. new_priceÍ∞íÏù¥ 9 Ïù¥ÌïòÏù¥Í≥† 
#      item_name Í∞íÏù¥ Chicken Salad Bowl Ïù∏
#      Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùÑ Ï∂îÏ∂ú
#-------------------------------------
Q27 = df.loc[
             (df['new_price'] <= 9) & 
             (df['item_name'] == 'Chicken Salad Bowl')
            ]
Q27.head()
```

```python
# Q28. dfÏùò new_price Ïª¨Îüº Í∞íÏóê Îî∞Îùº
#      Ïò§Î¶ÑÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ¶¨ÌïòÍ≥† indexÎ•º Ï¥àÍ∏∞Ìôî
#-------------------------------------
Q28 = df.sort_values('new_price').reset_index(drop=True)
Q28
```

```python
# Q29. dfÏùò item_name Ïª¨Îüº Í∞íÏ§ë 
#      Chips Ìè¨Ìï®ÌïòÎäî Í≤ΩÏö∞Ïùò Îç∞Ïù¥ÌÑ∞Î•º Ï∂úÎ†•
#-------------------------------------
Q29 = df.loc[df['item_name'].str.contains('Chips')]
Q29
```

```python
# Q30. dfÏùò ÏßùÏàòÎ≤àÏß∏ Ïª¨ÎüºÎßåÏùÑ Ìè¨Ìï®ÌïòÎäî Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Ï∂úÎ†•ÌïòÎùº
#-------------------------------------
Q30 = df.iloc[:,::2]
Q30
```
