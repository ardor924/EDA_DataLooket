#### 02 Filtering & Sorting
-----------------------------------------------------

```python
# Q20. 데이터로드
#-------------------------------------
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv')
```

```python
# Q21. quantity컬럼 값이 3인 데이터를 추출하여 첫 4행을 출력
#-------------------------------------
Q21 = df.loc[df['quantity'] == 3].head(4)
Q21
```

```python
# Q22. quantity컬럼 값이 3인 데이터를 추출하여 
#      index를 0부터 정렬하고 첫 6행을 출력
#-------------------------------------
Q22 = df.loc[df['quantity']==3].reset_index(drop=True).head(6)
Q22
```

```python
# Q23. quantity , item_price 두개의 컬럼으로 구성된
#      새로운 데이터 프레임을생성
#-------------------------------------
Q23 = df[['quantity','item_price']]

Q23
```

```python
# Q24. item_price 컬럼의 달러표시 문자를 제거하고
#      float 타입으로 저장하여 new_price 컬럼에 저장한뒤
#      앞의 5개출력
#-------------------------------------
df['new_price'] = df['item_price'].str[1:].astype(float)
df['new_price'].head()
```

```python
# Q25. new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출
#-------------------------------------
Q25 = df.loc[df['new_price'] <= 5]
cnt = len(df.loc[df['new_price'] <= 5])
display(Q25)
print(cnt)
```

```python
# Q26. item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하고
#      index 값을 초기화
#-------------------------------------

Q26 = df.loc[df['item_name'] == 'Chicken Salad Bowl'].reset_index(drop=True)
Q26 
```

```python
# Q27. new_price값이 9 이하이고 
#      item_name 값이 Chicken Salad Bowl 인
#      데이터 프레임을 추출
#-------------------------------------
Q27 = df.loc[
             (df['new_price'] <= 9) & 
             (df['item_name'] == 'Chicken Salad Bowl')
            ]
Q27.head()
```

```python
# Q28. df의 new_price 컬럼 값에 따라
#      오름차순으로 정리하고 index를 초기화
#-------------------------------------
Q28 = df.sort_values('new_price').reset_index(drop=True)
Q28
```

```python
# Q29. df의 item_name 컬럼 값중 
#      Chips 포함하는 경우의 데이터를 출력
#-------------------------------------
Q29 = df.loc[df['item_name'].str.contains('Chips')]
Q29
```

```python
# Q30. df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라
#-------------------------------------
Q30 = df.iloc[:,::2]
Q30
```
