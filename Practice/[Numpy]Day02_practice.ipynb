```python
#===============================================================
# Q16. 8x8 매트릭스를 만들어 체크보드 패턴으로 채우기
#===============================================================

# 문제풀이 :
#---------------------------------------------------------------
Q16 = np.arange((64),dtype=int).reshape(8,8)
Q16[:]        = "0"
Q16[1::2,::2] = "1"
Q16[::2,1::2] = "1"



# 해답 :
#-----------------------------------------------------------------
Q16
```

```python
#===============================================================
# Q17. 8x8 랜덤 행렬 정규화하기
#===============================================================

# 문제풀이 :
#---------------------------------------------------------------
np.random.seed(42)
arr = np.random.random(64).reshape(8,8)
mean = np.mean(arr)
std = np.std(arr)

Q17 = ((arr - mean) / std)


# 해답 :
#-----------------------------------------------------------------
Q17
```

```python
#===============================================================
# Q18. 색상을 4바이트(RGBA)로 나타내는 사용자 지정 dtype 생성
#===============================================================

# 문제풀이 :
#---------------------------------------------------------------
Q18 = np.dtype([
                    ("r", np.ubyte, 1),
                    ("g", np.ubyte, 1),
                    ("b", np.ubyte, 1),
                    ("a", np.ubyte, 1)
               ])

# 해답 :
#-----------------------------------------------------------------
Q18
```

```python
#===============================================================
# Q19. 1차원 배열(1~20)의 5~17의 요소를 음수로 전환
#===============================================================
Q19 = np.arange(1,20+1).reshape(4,5)
Q19[(5 < Q19) & (Q19 < 17)] *= -1
# 해답 :
#-----------------------------------------------------------------
Q19
```
